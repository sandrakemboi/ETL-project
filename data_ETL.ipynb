```python
#Extract data
population_data = pd.read_csv('Population.csv',sep= ",",quotechar = '"',engine='python',skiprows=4)
population_data.head(20)
```

```python
# Get list of countries
countries_data = pd.read_csv('un_member_iso_region.csv')
countries_data.info()
countries_data.head()
```

```python
corona_cases  = pd.read_csv('country_wise_latest.csv')
corona_cases.info()
```

```python
#clean data
def clean_Data(data):
    clean_pop_data = data[['Country Name','Country Code','2020']]
    #Filter to remain with UN countries
    un_countries = clean_pop_data [clean_pop_data['Country Code'].isin(countries_data['iso3'])]
    #Merge data with COVID-19 cases
    merged_data = pd.merge(un_countries,corona_cases,left_on = 'Country Name',right_on ='Country/Region', how='left')
    corona_un = merged_data.drop(['Deaths / 100 Cases','Recovered / 100 Cases','Deaths / 100 Recovered','Confirmed last week','1 week change','Country/Region'], axis = 1)
    return corona_un

UN_Corona =clean_Data(population_data)
```

```python

#Fill missing values
UN_Corona.fillna(0.0, inplace= True)
UN_Corona.info()
#Rename columns
UN_Corona = UN_Corona.rename(columns={'2020':'Population 2020'})
UN_Corona.info()
```
